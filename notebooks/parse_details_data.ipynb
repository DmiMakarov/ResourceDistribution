1. материал

Колонка количество изделий (обычно всегда 1) - означает, сколько изделия получится из этого компонента
Компонент - делается из указанного материала (материал - строчка, которая имеет только заполненную колонку "изделие")
Изделие с отсутсвующими другими значениеми - материал, который используетсся для создания компонента

2. Работа

Аналогично. Работа, для получения какого компонента она применяется
Компоненты для разных работ могут быть разные.
Получается, что может быть несколько работ, которые приводят в этот компонент. 
Проблема - как понять, какая операция соответсвует. Решение - по подразделению. Подразделение указано для каждой работы, но не для каждого материлала (потому что уникально для детали). 

То есть что надо сделать -> 
Сделать соответсвие изделий и базовых изделий, и изделий к подразделению по датафрейму материалов
В датафрейме работы сопоставить изделия к подразделениям и работам
По этим двум соответствиям для каждого изделия сделать дерево. Учесть, что операция, например, сборки, предполагает, что компонент дверь для изделия дверь, то есть в таком случае замыкает все связи на корне

```python
detail_to_base_detail: dict[str, list[dict]] = {}

current_base_detail: dict = {}
for _, detail_row in door_details.iterrows():

    if pd.isna(detail_row['Компонент']):
        material = detail_row['Изделие']
        name = material[material.find(' ') + 1:material.find('Сумма:') - 1]
        sum_ = float(material[material.rfind('Сумма:') + len('Сумма:'):material.find('Кол-во:') - 2].replace(',', '.'))
        current_base_detail = {'name': name, 'sum': sum_}

        continue

    current_base_detail['count'] = detail_row['Кол-во']

    if detail_row['Компонент'] in detail_to_base_detail:
        detail_to_base_detail[detail_row['Компонент']].append(current_base_detail)
    else:
        detail_to_base_detail[detail_row['Компонент']] = [current_base_detail]
```
